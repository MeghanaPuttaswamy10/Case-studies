```python
indicators = {"NY.GDP.PCAP.CD": "GDP per capita",
              "SP.DYN.IMRT.IN": "Infant mortality rate"}
df = wbdata.get_dataframe(indicators)
```

```python
print(df.isna().sum())
df = df.dropna()
df.count()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df['log_GDP_per_capita'] = np.log(df['GDP per capita'])
df['log_Infant_mortality_rate'] = np.log(df['Infant mortality rate'])
sns.scatterplot(x="log_GDP_per_capita", y="log_Infant_mortality_rate", data=df)

# Add a trendline using linear regression
#sns.regplot(x="GDP per capita", y="Infant mortality rate", data=df, scatter=False)

    # Set the title and axis labels
plt.title('Scatter plot')
plt.xlabel("GDP per capita (log scale)")
plt.ylabel("Infant mortality rate")

    # Display the plot
plt.show()
```

```python
plt.hist(df['log_GDP_per_capita'], density=True, bins=30)  # density=False would make counts
plt.ylabel('Density')
plt.xlabel('GDP per capita log')
```

```python
plt.hist(df['Infant mortality rate log'], density=True, bins=30)  # density=False would make counts
plt.ylabel('Density')
plt.xlabel('Infant mortality rate log')
```

```python
import statsmodels.api as sm
df = df.reset_index()
model = sm.formula.mixedlm("log_Infant_mortality_rate ~ log_GDP_per_capita*date + C(country)", df, groups=df["country"])
result = model.fit()
```

```python
result.summary()
```

```python
# Test the hypothesis that the relationship between GDP per capita and infant mortality rate is different among different countries over time
wald_test = result.wald_test_terms(["log_GDP_per_capita):date", "log_GDP_per_capita):C(country)"])
print(wald_test)
```
